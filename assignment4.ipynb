```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants
plt.rcParams['font.family'] = 'DejaVu Sans'
```

```python
# info
# O5 = 54000
# B5 = 15200
# A5 = 8310
# F5 = 6700
# G5 = 5660
# K5 = 4400
# M5 = 3200

# Define Wien's displacement constant
WIEN_CONST = 2.898e-3

# Define temperatures and star types
temperatures = [54000, 15200, 8310, 6700, 5660, 4400, 3200]
star_types = ['O5', 'B5', 'A5', 'F5', 'G5', 'K5', 'M5']

# Calculate maximum intensity wavelengths and store them in a list
max_int_wavelengths = []
for temp in temperatures:
    wavelength = WIEN_CONST / temp  # Wien's displacement law
    max_int_wavelengths.append(wavelength)

# Plot the calculated wavelengths
plt.plot(temperatures, max_int_wavelengths, 'ro')
plt.xlabel('Temperature (K)',fontsize=13)
plt.ylabel('Wavelength of maximum intensity (m)',fontsize=13)
plt.title("Wavelength of Maximum Intensity for Different Star Types vs Temperature", fontsize=13)
for i in range(len(star_types)):
    plt.annotate(star_types[i], (temperatures[i], max_int_wavelengths[i]), fontsize=13)
plt.show()


```

A note about units: recall from PHYS 208/245 that 1 Volt = 1 Joule/Coulomb. Electrons and protons give us a fundamental unit of electric charge, or elementary charge, $|q| = e = 1.602 \times 10^{−19}$ C. We can also calculate a fundamental unit of energy by calculating the change in kinetic energy of the elementary charge $e$ accelerated through an electric potential of 1 Volt. This new unit of energy is called the electron-Volt or eV. 1 eV = $1.602 \times 10^{−19}$ J.

```python


# Define constants
h = 4.136e-15 # Planck's constant in eV s
e = 1.602e-19 # Elementary charge in Coulombs
Al= 4.08
Au =5.1 # Work function in eV

# Define frequency range
nu = np.linspace(0, 3e15, 1000) # Hz

# Calculate maximum kinetic energy for each metal
Al_max = (h*nu - Al)
Au_max = (h*nu - Au)


# Plot the results
plt.plot(nu, Al_max, color = 'Black',label="Aluminum")
plt.plot(nu, Au_max,color = 'Gold',label="Gold")
plt.axvline(x=int(Al/h), color='gray', linestyle='--',label="Cutoff freq. Al")

plt.axvline(x=int(Au/h), color='gold', linestyle='--',label="Cutoff freq. Au")
plt.xlabel("Frequency (Hz)", fontsize=14)
plt.ylabel("Maximum Kinetic Energy (eV)", fontsize=14)
plt.legend(fontsize=13)
plt.title("Max KE vs Frequency", fontsize=14)
plt.show()
```
